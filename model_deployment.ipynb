```python
import os

models = {
    'yolov8n': 'runs/fruits_yolov8n/weights/best.pt',
    'yolov8m': 'runs/fruits_yolov8m/weights/best.pt',
    'yolo11l': 'runs/fruits_yolov11l/weights/best.pt',
    'faster_rcnn': 'models/faster_rcnn_fruits.pth'
}

for model_name, model_path in models.items():
    exists = os.path.exists(model_path)
    status = "✓" if exists else "✗"
    print(f"{status} {model_name}: {model_path}")
    if exists:
        size_mb = os.path.getsize(model_path) / (1024 * 1024)
        print(f"  {size_mb:.2f} MB")
```

```python
from huggingface_hub import HfApi, create_repo
import time

api = HfApi()
HF_USERNAME = "Danielx927"
REPO_ID = f"{HF_USERNAME}/fruit-detection-models"

print("Testing Hugging Face connection...")
try:
    api.whoami()
    print("✓ Connected to Hugging Face")
except Exception as e:
    print(f"✗ Connection failed: {e}")
    raise

print(f"\nCreating repository: {REPO_ID}")
create_repo(repo_id=REPO_ID, repo_type="model", exist_ok=True, private=False)
print("✓ Repository ready")

def upload_model(local_path, repo_path):
    size_mb = os.path.getsize(local_path) / (1024 * 1024)
    print(f"\nUploading {repo_path} ({size_mb:.2f} MB)...")
    
    start_time = time.time()
    try:
        api.upload_file(
            path_or_fileobj=local_path,
            path_in_repo=repo_path,
            repo_id=REPO_ID,
            repo_type="model"
        )
        elapsed = time.time() - start_time
        print(f"✓ Uploaded {repo_path} in {elapsed:.1f}s")
        return True
    except Exception as e:
        print(f"✗ Failed to upload {repo_path}: {e}")
        return False
```

```python
models_to_upload = [
    ('runs/fruits_yolov8n/weights/best.pt', 'yolov8n.pt'),
    ('runs/fruits_yolov8m/weights/best.pt', 'yolov8m.pt'),
    ('runs/fruits_yolov11l/weights/best.pt', 'yolov11l.pt'),
    ('models/faster_rcnn_fruits.pth', 'faster_rcnn.pth')
]

print("\nValidating files before upload...")
valid_models = []
for local_path, repo_path in models_to_upload:
    if os.path.exists(local_path):
        size_mb = os.path.getsize(local_path) / (1024 * 1024)
        print(f"✓ {repo_path}: {size_mb:.2f} MB")
        valid_models.append((local_path, repo_path))
    else:
        print(f"✗ Missing: {local_path}")

if not valid_models:
    print("\n✗ No valid models found to upload!")
else:
    print(f"\n{'='*60}")
    print(f"Starting upload of {len(valid_models)} models...")
    print(f"{'='*60}")
    
    success_count = 0
    for local_path, repo_path in valid_models:
        if upload_model(local_path, repo_path):
            success_count += 1
        time.sleep(1)  # Brief pause between uploads
    
    print(f"\n{'='*60}")
    print(f"Upload complete: {success_count}/{len(valid_models)} successful")
    print(f"{'='*60}")
    print(f"\nView at: https://huggingface.co/{REPO_ID}")
```
